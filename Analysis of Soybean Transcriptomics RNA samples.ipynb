```python
test_data.columns = test_data.columns.to_series().apply(lambda x: x.strip())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
features = ['Concentration', 'A260/A280', 'A260/A230']
```

```python
x = test_data.loc[:, features].values
```

```python
y = test_data.loc[:,['Target']].values
```

```python
x = StandardScaler().fit_transform(x)
```

```python
x
```

```python
y
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
```

```python
finalDf = pd.concat([principalDf, test_data[['Target']]], axis = 1)
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Projection on 1', fontsize = 15)
ax.set_ylabel('Projection on 2', fontsize = 15)
ax.set_title('Scatter plot Projection', fontsize = 20)

targets = ['Control_D1', 'Control_D7', 'Control_D14']
colors = ['r', 'g', 'b']
for Target, color in zip(targets,colors):
    indicesToKeep = finalDf['Target'] == Target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
